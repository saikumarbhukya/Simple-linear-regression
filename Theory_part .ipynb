### HeatMap

- It is used to find the correlation but in a graphical representation manner 

```python
data_d = data.iloc[: , 0:5]
```

```python
## Range of correlation = -1 to 1 
```

```python
data_d.head()
```

```python
data_d.corr()
```

```python
plt.figure(figsize = (13,7))
sns.heatmap(data_d.corr(),vmin=-1,vmax=1)
```

```python
plt.figure(figsize = (13,7))
sns.heatmap(data_d.corr(),vmin=-1,vmax=1,cmap = 'tab20')
```

```python
plt.figure(figsize = (13,7))
sns.heatmap(data_d.corr(),vmin=-1,vmax=1,annot=True,cmap = 'PiYG')
```

```python
data_d.corr()
```

```python
plt.figure(figsize = (13,7))
sns.heatmap(data_d.corr(),vmin=-1,vmax=1,annot=True,cmap = 'YlGnBu')
```

- default pearson correlation will be selected 

```python
plt.figure(figsize = (13,7))
sns.heatmap(data_d.corr(method = 'spearman'),vmin=-1,vmax=1,annot=True,cmap = 'YlGnBu',xticklabels=False,yticklabels=False)
```
